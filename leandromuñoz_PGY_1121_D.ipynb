```python


import random

personas = []

class Persona:
    def __init__(self, nif, nombre, edad):
        self.nif = nif
        self.nombre = nombre
        self.edad = edad

    def pertenece_union_europea(self):
        return random.choice([True, False])

def validar_nif(nif):

    return True

def grabar_persona():
    nif = input("Introduce el NIF: ")
    while not validar_nif(nif):
        print("El NIF es incorrecto. Inténtalo de nuevo.")
        nif = input("Introduce el NIF: ")

    nombre = input("Introduce el nombre: ")
    while len(nombre) <= 8:
        print("El nombre debe tener al menos 8 caracteres. Inténte de nuevo.")
        nombre = input("Introduce el nombre: ")

    edad = int(input("Introduce la edad: "))
    while edad <= 0:
        print("La edad debe ser mayor o igual a 0. Inténte de nuevo.")
        edad = int(input("Introduce la edad: "))

    persona = Persona(nif, nombre, edad)
    personas.append(persona)
    print("Persona grabada correctamente.")

def buscar_persona():
    nif = input("Introduce el NIF de la persona a buscar: ")
    for persona in personas:
        if persona.nif == nif:
            print("NIF: ", persona.nif)
            print("Nombre: ", persona.nombre)
            print("Edad: ", persona.edad)
            print("Pertenece a la Unión Europea: ", persona.pertenece_union_europea())
            return
    print("No se encontró ninguna persona con el NIF proporcionado.")

def imprimir_certificado():
    nombre_certificado = input("Introduce el nombre del certificado: ")
    nif_persona = input("Introduce el NIF de la persona: ")
    nombre_persona = input("Introduce el nombre de la persona: ")


    print("Certificado:", nombre_certificado)
    print("NIF:", nif_persona)
    print("Nombre de la persona:", nombre_persona)

def mostrar_menu():
    print("MENU")
    print("1. Grabar persona")
    print("2. Buscar persona")
    print("3. Imprimir certificado")
    print("4. Salir")

def ejecutar_opcion(opcion):
    if opcion == "1":
        grabar_persona()
    elif opcion == "2":
        buscar_persona()
    elif opcion == "3":
        imprimir_certificado()
    elif opcion == "4":
        print("Adios")
        exit()
    else:
        print("Opción erronea. Inténtalo de nuevo.")

def main():
    while True:
        mostrar_menu()
        opcion = input("Selecciona una opción: ")
        ejecutar_opcion(opcion)

if __name__ == "__main__":
    main()





































































```
